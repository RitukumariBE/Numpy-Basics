# Numpy student Marks Analytics
- In this project we are going to work on a dataset which in telling about the marks of student and we are going to analysis it and taking output from it

## Dataset Creation

```python
import numpy as np
```

### This dataset includes
- Their is 10 rows and the rows corresponds to the students.
- Their is 5 columns and columns corresponds to the subjects.

```python
# This array has 10 rows (students) and 5 columns (subjects)
marks = np.random.randint(1,100,50).reshape(10,5)
```

```python
marks.shape
```

```python
marks
```

## Understanding the Data


- To show the marks of the 1st student

```python
marks[0] # accessing the marks of student 1
```

- To check the marks of all students in the 2nd subject

```python
marks[:,1] # accessing the subject 2 marks of all students
```

- To check the last two subject marks of first three students

```python
marks[0:3,3:]
```

## Subject-wise Analysis
- In this section we are going to analysis subject wise performance using the statistical measures

### Average of each subject

```python
average1 = marks.mean(axis=0)
average1
```

```python
average1.max() # the hightest average among all the subjects
```

### Maximum of each subject

```python
marks.max(axis=0)
```

### Minimum of each subjecs

```python
marks.min(axis = 0)
```


- Highest average comes in the subject 1 with 60.4 . The 5th subject seems most difficult as the average comes out 27.3 . The performance is not looking balanced as one id=s 60 where other is 27 .
